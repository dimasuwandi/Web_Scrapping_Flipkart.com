```python
# https://www.kaggle.com/dimasuwandi/iphone-price-from-flipkartcom

from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
filename='dataScraping/iphoneFlipkart.csv'
f = open(filename,'a')
headers='Status,Product_Name,Memmory,Color,Price\n'
f.write(headers)
f.close()
```

```python
# for i in range(5):
pg = 1
while(True):
    my_url = f'https://www.flipkart.com/search?q=iphone&otracker=start&as-show=on&as=off&page={pg}'
    uClient = uReq(my_url)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, 'html.parser')
    
    containers = page_soup.findAll('div', {'class': 't-0M7P _2doH3V'})
    container = containers[0]

    lst = []
    raw = container.findAll('div', {'class':'_3liAhj'})
    for i in raw:
        lst.append(i.text)
        
    if len(lst)>0:
        print('page',pg)
        
        status = ''
        nma = ''
        warna = ''
        mem = ''
        harga = ''

        for j in range(len(lst)):
            nma = ' '.join(lst[j].split('(')[0].split(' ')[2:-1])
            if nma.split(' ')[0] == 'iPhone':
                try:
                    status = ' '.join(lst[j].split('(')[0].split(' ')[0:2])
                    warna = lst[j].split(',')[0].split('(')[len(lst[j].split(',')[0].split('('))-1].replace('PRODUCT)','')
                    mem = lst[j].split(',')[1].split(' ')[1]
                    lenPrice = len(lst[j].split('(')[2].split('₹'))
                    if lenPrice == 2:
                        harga = int(''.join(lst[j].split('(')[2].split('₹')[1].split('A')[0].replace(',','')))
                    elif lenPrice == 3:
                        harga = int(''.join(lst[j].split('(')[2].split('₹')[1]))

                    print(status,nma,warna,mem,harga)
                    
                    filename='dataScraping/iphoneFlipkart.csv'
                    f = open(filename,'a')
#                     f.write(str(status+','+nma+','+warna+','+mem+','+harga+'\n'))
                    f.write(status+','+nma+','+mem+','+warna+','+str(harga)+'\n')
                    f.close()

                except Exception as e:
#                     print(e)
                    continue

                
                
                    
    else:
        break
        f.close()
    
    pg = pg + 1
            



```
